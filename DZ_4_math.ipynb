1. Используя встроенные функции numpy (np.random.normal) сгенирируйте нормальное распределение c размерами выборки 10, 100, 1000 и 10000. Постройте гистограммы распределений. Рассчитайте среднее значение, медиану и моду для каждой выборке и сравните эти величины в пределах выборки.

```python
np.random.seed(15)
```

```python
samples10 = []
```

Выборка по 10

```python
for i in range(3000):
    sample = df.loc[:, 'capacity_mw'].sample(n=10).values
    samples10.append(sample)
samples10
```

Среднее

```python
mean_values = []
```

```python
for sample in samples10:
    mean_values.append(sample.mean())
```

```python
mean_values = np.array(mean_values)
```

```python
mean_values
```

```python
plt.hist(mean_values, bins=10)
plt.show()
```

Среднее значение сильно скачут, что говорит о слишком маленькой выборкой 

Медиана

```python
median_values=[]
```

```python
for sample in samples10:
    median_values.append(np.median(sample))
```

```python
plt.hist(median_values, bins=10)
plt.show()
```

График далек от графика нормального распределения

Мода

```python
mode_values=[]
```

```python
samples10r=np.around(samples10)
samples10r
```

```python
for sample in samples10:
    vals,counts = np.unique(sample, return_counts=True)
    index = np.argmax(counts)
    mode_values.append(vals[index])
```

```python
mode_values
```

```python
plt.hist(mode_values, bins=10)
plt.show()
```

$$f(x) = g e^{-x^2 + 6x 3}$$

Выразим неизвестный параметр g

$$g=f(x)*exp(x^2-6*x^3)$$

По формуле нормального распределния 

$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$

Найдем дисперсию M(X)

$$x^2-6*x^3=-{(x-a)^2}/{2*\sigma^2}$$

$a=x-\sqrt{(6*x^3-x^2)/2*\sigma^2}$

$a = M(X)$

$M(X)=a=x-\sqrt{(6*x^3-x^2)/2*\sigma^2}$

Найдем дисперсию D(X)

$\sigma^2=D(X).$

$D(X)=\sqrt{1/\sqrt{*pi}/g}$
