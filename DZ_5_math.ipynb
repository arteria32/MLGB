№1 Найти доверительный интервал для оценки математического ожидания $a$ нормального распределения с надежностью 0,95, зная выборочную среднюю $ \overline{x} = 75.12$, объем выборки $n=121$ и среднее квадратическое отклонение $\sigma=11$.
----


Воспользуемся формулой
$$T_{1,2} = \overline{X} \pm \frac{s_0}{\sqrt{n}} \cdot c_{\gamma}$$
Несмещенное среднее квадратичное отклонение 
$$s_0={\sqrt{n/(n-1)}*\sigma}$$

```python
import numpy as np
n=121
x=75.12
s=11
c=1/0.68
s0=np.sqrt(n/(n-1))*s
h=s0/np.sqrt(n)*c
T1=x-h
T2=x+h
print('Доверительный интервал(',T1,';',T2,')')
```

№2.С целью размещения рекламы опрошено 600 телезрителей, из которых данную передачу смотрят 170 человек. С доверительной вероятностью 0,91 найти долю телезрителей, охваченных рекламой в наихудшем случае.
-----

Для оценки неизвестной доли телезрителей используем формулу: 
$$w - t\sqrt{\frac{w(1-w)}{n}} < p < w + t\sqrt{\frac{w(1-w)}{n}}$$
где n - объём выборки, $w = \frac{170}{600} \approx 0.283$ - относительная частота события, $t = Ф^{-1} (\gamma/2) = Ф^{-1} (0.91/2) = 1.69$  (из таблицы функции Лапласа)

```python
N=600
k=170
w=k/N
p0=0.283-1.69*np.sqrt(w*(1-w)/N)
print('В худшем случае доля телезрителей равна',p0)
```
